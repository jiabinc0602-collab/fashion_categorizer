```python
#Importing libraries to test
import numpy as np
from neural_network import relu, softmax, one_hot, DeepNeuralNetwork
```

```python
#Testing one_hot function
y_test = np.array([0,1,2,0])
y_hot = one_hot(y_test, 3)
print(f"y_hot array: \n {y_hot}")
print(f"y_hot shape: {y_hot.shape}")
print(f"column 0 of y_hot: {y_hot[:,0]}")
```

```python
#Testing softmax function
Z_test = np.array([[10, 0, 100],
                   [10, 0, 100]])
A_test = softmax(Z_test)
print(f"Softmax Columns{np.sum(A_test, axis = 0)}")
print(f"Softmax col 0: {A_test[:,0]}")
```

```python
#Testing relu function
Z_relu = np.array([[-1, 2],
                   [0, -5]])
rel = relu(Z_relu)
print(f"Relu result: \n {rel}")
```

```python
X = np.random.randn(784, 10)
Y = np.eye(10)[:, :10]

#Initializing
model = DeepNeuralNetwork([784, 20, 10])

#Retrieving function returns
AL, caches = model.forward_propagation(X)
cost = model.compute_cost(AL, Y)
grads = model.backward_propagation(AL, Y, caches)
model.update_parameters(grads, 0.01)

#Checking to see if the functions work
print("Cost:", cost)
print("W1 change: ", model.params["W1"][0,0])
```
